#### Looping the URL List for Scrapping the Online Recipes

```python
rawRecipeLists = []
for url in urls:
    recipe = []
    scraper = scrape_me(url)
    recipe.append(scraper.title())
    recipe.append(scraper.ingredients())
    rawRecipeLists.append(recipe)
```

#### Save the List Containing the Scrapped Data into a File

```python
with open('rawRecipeLists.txt', 'w') as f:
    f.write(json.dumps(rawRecipeLists))
```

#### Open the File again for Checking

```python
with open('rawRecipeLists.txt', 'r') as f:
    rawRecipeLists = json.loads(f.read())
len(rawRecipeLists)
```
